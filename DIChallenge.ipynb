```python
# Two Pointers 
def printNumbers(start, end): 
    if start == end: 
        print(start)
    else: 
        if start < 0: 
            print("(" + str(start) + ")-", end = "")
        else: 
            print(str(start) + "-", end = "")
        if end < 0: 
            print("(" + str(end) + ")-", end = "")
        else: 
            print(end)

def printRangeWithHyphens(nums, start, end): 
    next = start 
    for num in nums: 
        if num < start: 
            continue 
        if num > end: 
            printNumbers(next, end)
            return 
        if num > next: 
            printNumbers(next, num - 1)
        next = num + 1

    if next <= end: 
        printNumbers(next, end)

printRangeWithHyphens(arr, 5, 12)
```
