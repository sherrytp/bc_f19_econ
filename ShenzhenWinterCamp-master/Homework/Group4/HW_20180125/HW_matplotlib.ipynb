```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Set axis labels
cross_x_1 = -3*np.pi/4
cross_x_2 = np.pi/4
cross_y_1 = np.sin(cross_x_1)
cross_y_2 = np.sin(cross_x_2)
ax.set(xticks=[cross_x_1,cross_x_2], yticks=[cross_y_1,cross_y_2])

x = np.linspace(-1*np.pi,np.pi,100)
y = np.sin(x)
label_y = r'sin'
z = np.cos(x)
label_z = r'cos'
ax.plot(x,y,linewidth=2,alpha=0.6,label=label_y,color='b')
ax.plot(x,z,linewidth=2,alpha=0.6,label=label_z,color='g')
ax.plot(cross_x_1,cross_y_1,'ro')
ax.plot(cross_x_2,cross_y_2,'ro')
ax.legend(loc='best')
plt.show()
```
