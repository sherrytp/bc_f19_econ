### Exercise 8

```r
# a. 
retaildata <- readxl::read_excel("data/retail.xlsx", skip = 1)
myts <- ts(retaildata[, "A3349588R"], 
           frequency = 12, start = c(1982, 4))
# Decomposition graph
myts %>% decompose(type = "additive") %>% 
  autoplot() + xlab("Year") 
myts %>% decompose(type="multiplicative") %>%
  autoplot() + xlab("Year") +
  ggtitle("Classical multiplicative decomposition")
```

- a. Why is multiplicative seasonality necessary? 

Because the comparison of simple additive and classical multiplicative shows the neccessarity, the addicative decomposition could not handle smoothly the remainders as stationary, where only multiplicative seasonality will leave out a roughly horizontal line around the 0 axis. 

Additionally, multiplicative method is preferred when the seasonal variations are changing proportional to the level of the series. So here, the seasonality is with changes not primarily constant. 

```r
myts2 <- hw(myts, seasonal = "multiplicative")
```

```r
# b. 
myts1 <- holt(y = myts, h = 25, damped = FALSE, level = c(80, 95), fan = FALSE,
  initial = c("optimal", "simple"), exponential = FALSE,
  alpha = NULL, beta = NULL, phi = 0.9, lambda = NULL,
  biasadj = FALSE, method = "multiplicative") 
# why damped is not working? 
myts1
```

```r
autoplot(myts1)
```

```r
# c. comparison of two models 
autoplot(myts) + 
    autolayer(myts1, series = "Holt Winter's forecasts", PI = FALSE) + 
    autolayer(myts2, series = "Multiplicative forecasts", PI = FALSE) + 
    xlab("Time") + 
    ylab("Retail") + 
    ggtitle("Retail Data A3349588R Forecasts") + 
    guides(color = guide_legend(title = "Forecast"))
# RMSE 
accuracy(myts1)
accuracy(myts2)
```

* c. Comparing only with the RMSE and the error term, myts2 of Holt-Winter's Multiplicative has the lower RMSE, which we prefer. 

```r
# d. 
checkresiduals(myts1)
checkresiduals(myts2)
```

```r
# e. 
myts.train <- window(myts, end = c(2010, 12))
myts.test <- window(myts, start = 2011)
autoplot(myts) + 
  autolayer(myts.train, series = "Train") + 
  autolayer(myts.test, series = "Test")
```

```r
# seasonal naive 
myts_snaive <- snaive(myts.train)
accuracy(myts_snaive, myts.test)
```

```r
# Holt's 
myts_holt <- holt(y = myts.train, damped = FALSE, level = c(80, 95), fan = FALSE,
  initial = c("optimal", "simple"), exponential = FALSE,
  alpha = NULL, beta = NULL, phi = NULL, lambda = NULL,
  biasadj = FALSE, method = "multiplicative") 
accuracy(myts_holt, myts.test)
```

```r
# Multiplicative 
myts_mul <- hw(myts.train, seasonal = "multiplicative")
accuracy(myts_mul, myts.test)
```

* e. No, seasonal naïve method still has the lowest RMSE among the three.
